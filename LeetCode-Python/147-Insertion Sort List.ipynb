```python
"""
时间复杂度 3624ms
"""

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None



class Solution:
    def insertionSortList(self, head: ListNode) -> ListNode:
        if not head:
            return head

        dummy_head = ListNode(None)
        dummy_head.next = head

        # 在前cnt个节点中寻找比node大的节点的前一个
        def find(cnt, node):
            pre = dummy_head
            cur = pre.next      # 错误这里cur=head，head可能改变，应该用dummy_head.next代替
            while cnt:
                if cur.val > node.val:
                    return pre
                pre = cur
                cur = cur.next
                cnt -= 1
            return None

        pre = dummy_head.next
        cur = pre.next
        cnt = 1
        while cur:
            t = find(cnt, cur)
            if t:
                # del cur node
                pre.next = cur.next
                # insert cur node into sorted linked list
                nxt = t.next
                cur.next = nxt
                t.next = cur
                # update pre, cur and cnt
                cur = pre.next
                cnt += 1
                continue

            pre = cur
            cur = cur.next
            cnt += 1
        return dummy_head.next
        
```

```python
"""如果cur节点已经大于前序pre节点，说明cur已经在合适位置，无需进入下面的循环（归并排序中的小技巧）。
能大幅缩短时间复杂度，从3000ms降到560ms"""


class Solution:
    def insertionSortList(self, head: ListNode) -> ListNode:
        if not head:
            return head

        dummy_head = ListNode(None)
        dummy_head.next = head

        pre = dummy_head.next
        cur = pre.next  # 当前待排序的节点
        while cur:
            if cur.val >= pre.val:
                pre = pre.next
            else:
                insert = dummy_head
                now = insert.next
                while now != cur and now.val <= cur.val:
                    insert = insert.next
                    now = insert.next
                if now != cur:  # now.val>cur.val
                    pre.next = cur.next  # cur will be removed
                    cur.next = now
                    insert.next = cur
                else:  # now == cur
                    pre = pre.next
            cur = pre.next
        return dummy_head.next
```

```python
"""
进一步在比较待排序节点和已排序节点过程中，减少一个指针
因为仅仅需要一个在链表中插入节点的位置，只要一个指针就够了
当p.next不符合要求时，p正好是可以插入的节点
这一修改也使时间复杂度降低到296ms，有点奇怪
"""
class Solution:
    def insertionSortList(self, head: ListNode) -> ListNode:
        if not head:
            return head

        dummy_head = ListNode(None)
        dummy_head.next = head

        pre = dummy_head.next
        cur = pre.next  # 当前待排序的节点
        while cur:
            if cur.val >= pre.val:
                pre = pre.next
            else:
                p = dummy_head
                while p.next != cur and p.next.val <= cur.val:
                    p = p.next
                if p.next != cur:  # p.next.val > cur.val
                    pre.next = cur.next  # cur will be removed
                    cur.next = p.next
                    p.next = cur
                else:  # pre.next == cur
                    pre = pre.next
            cur = pre.next
        return dummy_head.next
```

```python
"""
每次将待排序节点和已排序节点比较过程中，都是从头开始，这一点可以进一步优化
加入遍历指针p遍历到的已排序元素 大于 待排序节点，这时才需要从头开始。
否则可以从上次p节点开始据需遍历
这一优化使得时间复杂度为200ms
"""
class Solution:
    def insertionSortList(self, head: ListNode) -> ListNode:
        if not head:
            return head

        dummy_head = ListNode(None)
        dummy_head.next = head

        p = dummy_head
        pre = dummy_head.next
        cur = pre.next  # 当前待排序的节点
        while cur:
            if cur.val >= pre.val:
                pre = pre.next
            else:
                if p.val and p.val > cur.val:
                    p = dummy_head
                while p.next != cur and p.next.val <= cur.val:
                    p = p.next
                if p.next != cur:  # p.next.val > cur.val
                    pre.next = cur.next  # cur will be removed
                    cur.next = p.next
                    p.next = cur
                else:  # pre.next == cur
                    pre = pre.next
            cur = pre.next
        return dummy_head.next
```
