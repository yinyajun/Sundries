给定两个非空链表来代表两个非负整数。数字最高位位于链表开始位置。它们的每个节点只存储单个数字。将这两数相加会返回一个新的链表。

 

你可以假设除了数字 0 之外，这两个数字都不会以零开头。

进阶:

如果输入链表不能修改该如何处理？换句话说，你不能对列表中的节点进行翻转。

示例:

输入: (7 -> 2 -> 4 -> 3) + (5 -> 6 -> 4)

输出: 7 -> 8 -> 0 -> 7


```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        new_l1 = self.reverse(l1)
        new_l2 = self.reverse(l2)
        ret = self.add(new_l1, new_l2)
        return self.reverse(ret)
        
    
    def reverse(self, head):
        if head is None or head.next is None:
            return head
        
        dummy_head = ListNode(None)
        dummy_head.next = head
        
        pre = dummy_head
        cur = head
        
        while cur:
            post = cur.next
            cur.next = pre
            pre = cur
            cur = post
            
        head.next = None
        return pre
    
    def add(self, l1, l2):
        carry = 0
        dummy_head = ListNode(None)
        cur = dummy_head
        while l1 or l2 or carry:
            x = l1.val if l1 else 0
            y = l2.val if l2 else 0
            s = x + y + carry
            b = s % 10 
            carry = s // 10
            cur.next = ListNode(b)
            cur = cur.next
            
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next
        return dummy_head.next
        
            
            
        
```

```python
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        first = dummy_l1 = ListNode(None)
        dummy_l1.next = l1
        second = dummy_l2 = ListNode(None)
        dummy_l2.next = l2

        carry = 0

        m = self.length(l1)
        n = self.length(l2)

        def recursive(first, second):  # m>n
            nonlocal carry, m, n

            if m == 0:
                return None

            first = first.next
            if m == n:
                second = second.next
                n -= 1
            m -= 1

            child_head = recursive(first, second)

            x = first.val if first and first.val else 0
            y = second.val if second and second.val else 0

            s = x + y + carry
            b = s % 10
            carry = s // 10

            tmp = child_head
            child_head = ListNode(b)
            child_head.next = tmp
            return child_head

        if m >= n:
            new_head = recursive(first, second)
        else:
            m, n = n, m
            new_head = recursive(second, first)
        return new_head

    def length(self, head):
        cnt = 0
        if not head:
            return cnt
        while head:
            head = head.next
            cnt += 1
        return cnt
```
