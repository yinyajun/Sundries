## 生成训练集

```python
train_x = []
train_y = []
for i in range(15):
    x = data.iloc[:, i:i + 9]
    x.columns = np.array(
        range(9))  #notice if we don't set columns name, it will have different columns name in each iteration
    #     d=data.iloc[:,-2:]
    #     d.columns=[9,10]
    #     x=x.join(d)
    y = data.iloc[:, i + 9]
    y.columns = np.array(range(1))
    train_x.append(x)
    train_y.append(y)
```

```python
# 如何合并？merge,join或concat
# merge用于列，join用于index，concat相当于union或者outer join（axis=1）
a = data.iloc[:, -2:]
b = data.iloc[:, :9]
# pd.concat([a,b],axis=1).head()
a.join(b).head()
```

```python
train_x=pd.concat(train_x)
train_y=pd.concat(train_y)  
train_x.head(10)
```

```python
# 将含有负值的记录删除
cond1 = train_x.apply(lambda x: x > 0).all(axis=1)
x = train_x[cond1]
y = train_y[cond1]
cond2 = y.apply(lambda x: x > 0)
x = x[cond2]
y = y[cond2]
train_x = x
train_y = y
len(train_y)
```

```python
# 总共少的训练样本
3600-2886
```

```python
# 标准化
class standardScaler(object):
    def __init__(self, train_x):
        m_ = train_x.apply(lambda x:x.mean())
        s_ = train_x.apply(lambda x:x.std())
        self.m_ = np.array(m_, float)
        self.s_ = np.array(s_, float)
    
    def fit(self, df):
        df = df.sub(self.m_).divide(self.s_)
        return df
    
ss = standardScaler(train_x)
train_x = ss.fit(train_x)
train_x.head()
```

```python
train_x = np.array(train_x,float)
train_y = np.array(train_y,float)
# add bias to train dataset
train_x = np.concatenate((np.ones((train_x.shape[0],1)),train_x),axis=1)
train_x
```

```python
train_y
```

## 线性回归

```python
# 初始化
w = np.random.normal(0,0.001,len(train_x[0]))
lr = 1.0
epoch_num = 20000
```

```python
# 训练, AdaGrad
loss_history=[]
sum_grad = np.zeros(len(train_x[0]))
for i in range(epoch_num):
    y_hat = np.dot(train_x, w) # MxN * Nx1 = Mx1
    loss = ((train_y - y_hat) ** 2).mean()
    grad = np.dot(train_x.transpose(), y_hat - train_y) # NxM * Mx1 = Nx1
    sum_grad += grad ** 2
    ada_lr = lr / np.sqrt(sum_grad)
    w -= ada_lr * grad
    
    loss_history.append(loss)
    if i % 1000 == 0:
        print("Step: {0}, Loss: {1}".format(i, loss))
```
